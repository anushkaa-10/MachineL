```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df['species'].unique()
```

```python
df.isnull().sum()
```

```python
wout_setosa=df[df['species']!='setosa']
```

```python
wout_setosa.head()
```

```python
wout_setosa['species']=wout_setosa['species'].map({'versicolor':0,'virginica':1})
```

```python
wout_setosa.head()
```

```python
## split into dependent and independent features
X=wout_setosa.iloc[:,:-1]
y=wout_setosa.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
# Corrected the order of y_train and y_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV

# Adjust the parameter grid to use compatible solver and penalty combinations:
# 1. Remove 'elasticnet' from penalty when using 'lbfgs' solver.
# 2. Use 'saga' solver with 'elasticnet' and define a range for 'l1_ratio'.
# 3. Ensure 'liblinear' and 'saga' are used with compatible penalties.
parameter = {
    'penalty': ['l1', 'l2'],  # Penalties for liblinear and saga
    'C': [1, 2, 3, 4, 5, 6, 8, 10, 20, 30, 40, 50],
    'max_iter': [100, 200, 300],  # Adjust max_iter values
    'solver': ['liblinear', 'lbfgs', 'saga']
}

# Define a separate parameter grid for lbfgs to avoid the penalty conflict
parameter_lbfgs = {
    'penalty': ['l2'],  # Only l2 penalty for lbfgs
    'C': [1, 2, 3, 4, 5, 6, 8, 10, 20, 30, 40, 50],
    'max_iter': [100, 200, 300],  # Adjust max_iter values
    'solver': ['lbfgs']
}

# Create two GridSearchCV objects, one for each parameter grid
classifier_regressor_liblinear_saga = GridSearchCV(classifier, param_grid=parameter, scoring='accuracy', cv=5)
classifier_regressor_lbfgs = GridSearchCV(classifier, param_grid=parameter_lbfgs, scoring='accuracy', cv=5)

# Fit the models separately
classifier_regressor_liblinear_saga.fit(X_train, y_train)
classifier_regressor_lbfgs.fit(X_train, y_train)

# Compare the results and choose the best model
# ...
```

```python
print(classifier_regressor_liblinear_saga.best_params_)
```

```python
print(classifier_regressor_lbfgs.best_params_)
```

```python
print(classifier_regressor_lbfgs.best_score_)
```

```python
print(classifier_regressor_liblinear_saga.best_score_)
```

```python
y_pred=classifier_regressor_liblinear_saga.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
y_pred2 = classifier_regressor_lbfgs.predict(X_test)
```

```python
score2 = accuracy_score(y_pred2,y_test)
print(score2)
```

```python
print(classification_report(y_pred2,y_test))
```
